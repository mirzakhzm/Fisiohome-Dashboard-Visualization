DATA TYPE CORRECTION

```python
df.dtypes
```

```python
def extract_first_number(usia_str):
    import re
    match = re.search(r'\d+', usia_str)
    return int(match.group()) if match else None

df['Usia'] = df['Usia'].apply(extract_first_number)
```

```python
df['Usia'] = df['Usia'].fillna(0).astype(int)
df['Tanggal Visit'] = pd.to_datetime(df['Tanggal Visit'])
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```
