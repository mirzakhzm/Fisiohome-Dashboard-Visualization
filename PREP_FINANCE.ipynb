```python
def clean_discount(discount):
    if isinstance(discount, str):
        if 'Rp' in discount:
            # Bersihkan simbol 'Rp' dan koma
            return int(discount.replace('Rp', '').replace(',', ''))
    return discount

df['Discount'] = df['Discount'].apply(clean_discount)
```

```python
def clean_price(price):
    if isinstance(price, str):
        if 'Rp' in price:
            # Bersihkan simbol 'Rp' dan koma
            return int(price.replace('Rp', '').replace(',', ''))
    return price


df['Price'] = df['Price'].apply(clean_price)
```

```python
def clean_transaction(transaction):
    if isinstance(transaction, str):
        if 'Rp' in transaction:
            # Bersihkan simbol 'Rp' dan koma
            return int(transaction.replace('Rp', '').replace(',', ''))
    return transaction


df['Total Transaction'] = df['Total Transaction'].apply(clean_transaction)
```

```python
def clean_income(income):
    if isinstance(income, str):
        if 'Rp' in income:
            # Bersihkan simbol 'Rp' dan koma
            return int(income.replace('Rp', '').replace(',', ''))
    return income


df['Nett Income'] = df['Nett Income'].apply(clean_income)
```

```python
def clean_wallet(wallet):
    if isinstance(wallet, str):
        # Bersihkan simbol 'Rp', koma, titik, dan spasi
        cleaned_wallet = wallet.replace('Rp', '').replace(',', '').replace('.', '').strip()
        if cleaned_wallet.isdigit():
            return int(cleaned_wallet)
    return wallet

df['Wallet Mitra'] = df['Wallet Mitra'].apply(clean_wallet)
```

```python
cols = ['Price', 'Wallet Mitra', 'Discount','Nett Income','Total Transaction']
df[cols] = df[cols].apply(lambda x: pd.to_numeric(x, errors='coerce')).fillna(0).astype(int)
```
